```python
import plotly.express as px
import plotly.io as pio
pio.renderers.default = "jpeg"
```

```python
df = px.data.tips()
df.head()
```

```python
fig = px.bar(df, x="sex", y="total_bill", color="smoker", barmode="group")
fig.show()
```

```python
fig = px.bar(df, x="sex", y="total_bill", color="day", barmode="group")
fig.show()
```

```python
fig = px.scatter()
fig = px.scatter(df, x="total_bill", y="tip", color="sex", facet_col="smoker")
fig.show()
```

```python
import statsmodels

fig = px.scatter()
fig = px.scatter(df, x="total_bill", y="tip", 
color="sex", facet_col="smoker", trendline="ols")
fig.show()
```

```python
fig=px.scatter(data_frame=df, x="total_bill", y="tip", facet_row="time", facet_col="day")
fig.show()
```

```python
fig=px.histogram(data_frame=df, x="tip", marginal="rug")
fig.show()
```
